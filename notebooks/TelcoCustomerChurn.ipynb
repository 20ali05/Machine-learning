```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
```

```python
df = pd.read_csv("Telco_customer_churn.csv")
```

```python
gender_counts = df['Gender'].value_counts()

plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'orange'])
plt.title('Customer Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
gender_counts = df["Gender"].value_counts()

plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Customer Distribution by Gender')
plt.show()
```

```python
city_list = df["City"]
citys= []
for i in city_list:
    if i in citys:
        pass
    else:
        citys.append(i)
print(citys)
print(len(citys))
```

```python
city_counts = df["City"].value_counts()


plt.bar(city_counts.index, city_counts.values)
plt.title('Customer Distribution by city')
plt.xlabel('citys')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df['Monthly Charges'], bins=30)
plt.title('Distribution of Monthly Charges')
plt.xlabel('Monthly Charges')
plt.ylabel('Frequency')
plt.show()
```

```python
churn_counts = df['Churn Label'].value_counts()

plt.pie(churn_counts.values, labels=churn_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Churn Distribution')
plt.show()
```

```python
plt.scatter(df['Monthly Charges'], df['Total Charges'], color='purple')
plt.title('Monthly Charges vs Total Charges')
plt.xlabel('Monthly Charges')
plt.ylabel('Total Charges')
plt.show()
```

```python
churn_reasons_counts = df['Churn Reason'].value_counts()

plt.bar(churn_reasons_counts.index, churn_reasons_counts.values, color='teal')
plt.xticks(rotation=90)
plt.title('Distribution of Churn Reasons')
plt.xlabel('Reason')
plt.ylabel('Count')
plt.show()
```
